## Cơ Sở Lý Thuyết

### 1. Tìm Kiếm Mù
Tìm kiếm mù (Uninformed Search) là các thuật toán tìm kiếm không sử dụng thông tin heuristic, duyệt qua không gian trạng thái theo cách có hệ thống, dựa trên cấu trúc bài toán.

### 2. Tìm Kiếm theo Chiều Rộng (BFS)
- **Nguyên lý**: Khám phá tất cả các nút ở mức hiện tại trước khi chuyển sang mức tiếp theo.
- **Cấu trúc dữ liệu**: Hàng đợi (Queue).
- **Đặc điểm**:
  - Trong đồ thị không trọng số, BFS đảm bảo tìm đường đi ngắn nhất (theo số cạnh).
  - Trong đồ thị có trọng số, BFS không đảm bảo đường đi có tổng trọng số nhỏ nhất, vì nó không xem xét trọng số của các cạnh.

### 3. Tìm Kiếm theo Chiều Sâu (DFS)
- **Nguyên lý**: Khám phá một nhánh đến tận cùng trước khi quay lại thử nhánh khác.
- **Cấu trúc dữ liệu**: Ngăn xếp (Stack) hoặc đệ quy.
- **Đặc điểm**:
  - DFS không đảm bảo đường đi ngắn nhất trong cả đồ thị không trọng số và có trọng số.
  - Kết quả phụ thuộc vào thứ tự duyệt các nút kề.

### 4. Đồ Thị Không Trọng Số và Có Trọng Số
- **Đồ thị không trọng số**: Các cạnh có trọng số ngầm định bằng 1. BFS tối ưu theo số cạnh.
- **Đồ thị có trọng số**: Các cạnh có trọng số khác nhau (ví dụ: chi phí, khoảng cách). BFS và DFS không được thiết kế để tối ưu hóa tổng trọng số, dẫn đến kết quả không tối ưu.

### 5. Bài Toán 
Cho các đồ thị vô hướng (không trọng số và có trọng số) biểu diễn mạng lưới hoặc mê cung. Nhiệm vụ là tìm đường đi từ nút bắt đầu (**S**) đến nút đích (**G** hoặc **H**) bằng BFS và DFS, thông qua minh họa thủ công và lập trình với chú thích chi tiết.

5. Lấy F, không có kề mới: Ngăn xếp rỗng, quay lại S.
6. Từ S, thêm B: Ngăn xếp = [(B, [S, B])], Đã thăm = {S, A, D, F, B}
7. Lấy B, thêm E: Ngăn xếp = [(E, [S, B, E])], Đã thăm = {S, A, D, F, B, E}

### Phần 3: Triển Khai Mã Python

#### BFS trên Đồ Thị Không Trọng Số
Chạy ô mã dưới đây để kiểm chứng BFS trên **Đồ thị mẫu 1**.

````python
from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 1 (không trọng số)
graph1 = {
    'S': ['A', 'B', 'C'],
    'A': ['S', 'D'],
    'B': ['S', 'E'],
    'C': ['S'],
    'D': ['A', 'F'],
    'E': ['B', 'G'],
    'F': ['D'],
    'G': ['E']
}

# Chạy BFS và in kết quả
path = bfs(graph1, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 1:", path)
````

#### DFS trên Đồ Thị Không Trọng Số
Chạy ô mã dưới đây để kiểm chứng DFS trên **Đồ thị mẫu 1**.

````python
def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path
    
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 1 (không trọng số)
graph1 = {
    'S': ['A', 'B', 'C'],
    'A': ['S', 'D'],
    'B': ['S', 'E'],
    'C': ['S'],
    'D': ['A', 'F'],
    'E': ['B', 'G'],
    'F': ['D'],
    'G': ['E']
}

# Chạy DFS và in kết quả
path = dfs(graph1, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 1:", path)
````

#### BFS trên Đồ Thị Có Trọng Số
Chạy ô mã dưới đây để kiểm chứng BFS trên **Đồ thị mẫu 5**.

````python
from collections import deque

def bfs_weighted(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu, đường đi, và tổng trọng số
    queue = deque([(start, [start], 0)])
    # Khởi tạo tập hợp các nút đã thăm
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút, đường đi, và tổng trọng số từ đầu hàng đợi
        (node, path, total_weight) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path, total_weight
        # Duyệt qua các nút kề và trọng số tương ứng
        for neighbor, weight in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề, đường đi mới, và tổng trọng số cập nhật vào hàng đợi
                queue.append((neighbor, path + [neighbor], total_weight + weight))
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 5 (có trọng số)
graph5 = {
    'S': [('A', 1), ('B', 2), ('C', 3)],
    'A': [('S', 1), ('D', 4), ('G', 3)],
    'B': [('S', 2), ('G', 5)],
    'C': [('S', 3)],
    'D': [('A', 4), ('G', 6)],
    'G': [('B', 5), ('D', 6), ('A', 3)]
}

# Chạy BFS và in kết quả
path, weight = bfs_weighted(graph5, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 5:", path, "Tổng trọng số:", weight)
````

#### DFS trên Đồ Thị Có Trọng Số
Chạy ô mã dưới đây để kiểm chứng DFS trên **Đồ thị mẫu 5**.

````python
def dfs_weighted(graph, start, goal, visited=None, path=None, total_weight=0):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path, total_weight
    
    # Duyệt qua các nút kề và trọng số tương ứng
    for neighbor, weight in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi và tổng trọng số mới
            new_path, new_weight = dfs_weighted(graph, neighbor, goal, visited, path + [neighbor], total_weight + weight)
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path, new_weight
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 5 (có trọng số)
graph5 = {
    'S': [('A', 1), ('B', 2), ('C', 3)],
    'A': [('S', 1), ('D', 4), ('G', 3)],
    'B': [('S', 2), ('G', 5)],
    'C': [('S', 3)],
    'D': [('A', 4), ('G', 6)],
    'G': [('B', 5), ('D', 6), ('A', 3)]
}

# Chạy DFS và in kết quả
path, weight = dfs_weighted(graph5, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 5:", path, "Tổng trọng số:", weight)
````

## Bài Tập lập trình

#### Bài Tập Lập Trình: BFS trên Đồ Thị Mẫu 2
Chạy BFS trên **Đồ thị mẫu 2** với chú thích chi tiết.

````python
# code here
from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 2 (có chu trình, không trọng số)
# Dựa trên sơ đồ: S->A, S->B, A->B, A->C, B->D, C->D, D->G
graph2 = {
    'S': ['A', 'B'],
    'A': ['B', 'C'],
    'B': ['D'],
    'C': ['D'],
    'D': ['G'],
    'G': []
}

# Chạy BFS và in kết quả
path = bfs(graph2, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 2:", path)
````

#### Bài Tập Lập Trình: DFS trên Đồ Thị Mẫu 2
Chạy DFS trên **Đồ thị mẫu 2** với chú thích chi tiết.

````python
# code here
def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    print(f"Đang thăm nút: {start}, Đường đi hiện tại: {path}")
    
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        print(f"Đã tìm thấy đích {goal}!")
        return path
    
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        print(f"Kiểm tra nút kề: {neighbor}")
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            print(f"Nút {neighbor} chưa được thăm, tiếp tục DFS")
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
        else:
            print(f"Nút {neighbor} đã được thăm, bỏ qua")
    
    print(f"Backtrack từ nút {start}")
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 2 (có chu trình, không trọng số)
graph2 = {
    'S': ['A', 'B'],
    'A': ['B', 'C'],
    'B': ['D'],
    'C': ['D'],
    'D': ['G'],
    'G': []
}
# Chạy DFS và in kết quả
path = dfs(graph2, 'S', 'G')
print("-" * 40)
print("Đường đi DFS trên Đồ thị mẫu 2:", path)
````

### Sửa mã BFS để đếm số nút đã thăm trên **Đồ thị mẫu 4**. In ra kết quả và đảm bảo mã có chú thích đầy đủ.

````python
# code here
from collections import deque

def bfs_count_visited(graph, start, goal):
    """
    Thực hiện BFS và đếm số nút đã thăm
    
    Args:
        graph: Đồ thị dạng dictionary
        start: Nút bắt đầu
        goal: Nút đích
    
    Returns:
        tuple: (đường đi, số nút đã thăm)
    """
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path, len(visited)
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    
    # Trả về None và số nút đã thăm nếu không tìm thấy đường đi
    return None, len(visited)

# Định nghĩa Đồ thị mẫu 4 (dạng lưới, không trọng số)
# Các cạnh: S-A, S-C, A-B, A-D, B-E, C-D, C-F, D-E, E-G, F-G
graph4 = {
    'S': ['A', 'C'],      # S kết nối với A và C
    'A': ['S', 'B', 'D'], # A kết nối với S, B và D
    'B': ['A', 'E'],      # B kết nối với A và E
    'C': ['S', 'D', 'F'], # C kết nối với S, D và F
    'D': ['A', 'C', 'E'], # D kết nối với A, C và E
    'E': ['B', 'D', 'G'], # E kết nối với B, D và G
    'F': ['C', 'G'],      # F kết nối với C và G
    'G': ['E', 'F']       # G kết nối với E và F
}

# Chạy BFS với đếm số nút đã thăm và in kết quả
path, visited_count = bfs_count_visited(graph4, 'S', 'G')
print("=== KẾT QUẢ BFS TRÊN ĐỒ THỊ MẪU 4 ===")
print(f"Đường đi từ S đến G: {path}")
print(f"Số nút đã thăm: {visited_count}")
print(f"Tổng số nút trong đồ thị: {len(graph4)}")
````

````python
# code here
# 1. Viết mã Python để chạy BFS và DFS trên **Đồ thị mẫu 6** và **Đồ thị mẫu 7**. Định nghĩa đồ thị dưới dạng từ điển và thêm chú thích chi tiết.
from collections import deque

def bfs(graph, start, end):
    """
    Thuật toán tìm kiếm theo chiều rộng (BFS)
    Args:
        graph: đồ thị dưới dạng từ điển
        start: đỉnh bắt đầu
        end: đỉnh kết thúc
    Returns:
        tuple: (đường đi, danh sách thứ tự duyệt)
    """
    queue = deque([(start, [start])])  # Queue chứa (đỉnh hiện tại, đường đi)
    visited = set()
    order = []  # Thứ tự duyệt các đỉnh
    
    while queue:
        node, path = queue.popleft()
        
        if node in visited:
            continue
            
        visited.add(node)
        order.append(node)
        
        if node == end:
            return path, order
            
        # Thêm các đỉnh kề vào queue
        for neighbor in sorted(graph.get(node, [])):
            if neighbor not in visited:
                queue.append((neighbor, path + [neighbor]))
    
    return None, order

def dfs(graph, start, end, visited=None, path=None, order=None):
    """
    Thuật toán tìm kiếm theo chiều sâu (DFS) - đệ quy
    Args:
        graph: đồ thị dưới dạng từ điển
        start: đỉnh bắt đầu
        end: đỉnh kết thúc
        visited: tập các đỉnh đã thăm
        path: đường đi hiện tại
        order: thứ tự duyệt các đỉnh
    Returns:
        tuple: (đường đi, danh sách thứ tự duyệt)
    """
    if visited is None:
        visited = set()
    if path is None:
        path = []
    if order is None:
        order = []
    
    visited.add(start)
    path.append(start)
    order.append(start)
    
    if start == end:
        return path.copy(), order
    
    # Duyệt các đỉnh kề theo thứ tự alphabet
    for neighbor in sorted(graph.get(start, [])):
        if neighbor not in visited:
            result_path, result_order = dfs(graph, neighbor, end, visited, path, order)
            if result_path:
                return result_path, result_order
    
    path.pop()  # Backtrack
    return None, order

# Đồ thị mẫu 6: Có trọng số, nhiều đường đi
# Lưu ý: Vì BFS/DFS không xử lý trọng số, ta chỉ lưu các đỉnh kề
graph6 = {
    'S': ['A', 'C'],
    'A': ['B', 'D'],
    'B': ['E'],
    'C': ['D', 'F'],
    'D': ['E'],
    'E': ['H'],
    'F': ['G'],
    'G': ['H'],
    'H': []
}

# Thông tin trọng số cho tham khảo (không sử dụng trong BFS/DFS)
weights6 = {
    ('S', 'A'): 2, ('S', 'C'): 5, ('A', 'B'): 3, ('A', 'D'): 4,
    ('B', 'E'): 6, ('C', 'D'): 7, ('C', 'F'): 9, ('D', 'E'): 8,
    ('E', 'H'): 10, ('F', 'G'): 12, ('G', 'H'): 15
}

# Đồ thị mẫu 7: Không trọng số, mật độ cao
graph7 = {
    'S': ['A', 'D', 'E'],
    'A': ['B', 'D', 'E'],
    'B': ['C', 'E', 'F'],
    'C': ['F', 'G'],
    'D': ['E'],
    'E': ['F', 'H'],
    'F': ['G', 'H'],
    'G': ['H'],
    'H': []
}

def run_algorithms(graph, graph_name, start='S', end='H'):
    """
    Chạy cả BFS và DFS trên đồ thị và in kết quả
    """
    print(f"\n=== {graph_name} ===")
    print(f"Tìm đường từ {start} đến {end}")
    
    # Chạy BFS
    print("\n--- BFS (Breadth-First Search) ---")
    bfs_path, bfs_order = bfs(graph, start, end)
    print(f"Thứ tự duyệt: {' -> '.join(bfs_order)}")
    if bfs_path:
        print(f"Đường đi tìm được: {' -> '.join(bfs_path)}")
    else:
        print("Không tìm thấy đường đi")
    
    # Chạy DFS
    print("\n--- DFS (Depth-First Search) ---")
    dfs_path, dfs_order = dfs(graph, start, end)
    print(f"Thứ tự duyệt: {' -> '.join(dfs_order)}")
    if dfs_path:
        print(f"Đường đi tìm được: {' -> '.join(dfs_path)}")
    else:
        print("Không tìm thấy đường đi")

# Chạy thuật toán trên cả hai đồ thị
if __name__ == "__main__":
    print("THUẬT TOÁN TÌM KIẾM BFS VÀ DFS")
    print("=" * 50)
    
    # Đồ thị mẫu 6
    run_algorithms(graph6, "Đồ thị mẫu 6: Có trọng số, nhiều đường đi")
    
    # Đồ thị mẫu 7
    run_algorithms(graph7, "Đồ thị mẫu 7: Không trọng số, mật độ cao")
    
    print("\n" + "=" * 50)
    print("GHI CHÚ:")
    print("- BFS tìm đường đi ngắn nhất về số cạnh")
    print("- DFS có thể tìm được đường đi khác với BFS")
    print("- Đồ thị 6 có trọng số nhưng BFS/DFS không xử lý trọng số")
    print("- Đồ thị 7 có mật độ cạnh cao, nhiều đường đi khác nhau")
````

````python
# code here
# 2. Sửa mã BFS để lưu và in tất cả các đường đi từ S đến H trên **Đồ thị mẫu 7** (không trọng số). Đảm bảo mã có chú thích đầy đủ.
from collections import deque

def bfs_all_paths(graph, start, target):
    """
    Tìm tất cả các đường đi từ start đến target sử dụng BFS
    
    Args:
        graph: Dictionary biểu diễn đồ thị dạng adjacency list
        start: Đỉnh bắt đầu
        target: Đỉnh đích
    
    Returns:
        List chứa tất cả các đường đi từ start đến target
    """
    # Queue lưu trữ các đường đi (mỗi đường đi là một list các đỉnh)
    queue = deque([[start]])
    all_paths = []  # Danh sách lưu tất cả đường đi tìm được
    
    while queue:
        # Lấy đường đi hiện tại từ queue
        current_path = queue.popleft()
        current_node = current_path[-1]  # Đỉnh cuối của đường đi hiện tại
        
        # Nếu đã đến đích, lưu đường đi
        if current_node == target:
            all_paths.append(current_path)
            continue
        
        # Duyệt qua tất cả đỉnh kề
        for neighbor in graph.get(current_node, []):
            # Kiểm tra để tránh chu trình (không quay lại đỉnh đã đi qua)
            if neighbor not in current_path:
                # Tạo đường đi mới bằng cách thêm đỉnh kề vào đường đi hiện tại
                new_path = current_path + [neighbor]
                queue.append(new_path)
    
    return all_paths

def print_all_paths(paths, start, target):
    """
    In tất cả các đường đi tìm được
    
    Args:
        paths: Danh sách các đường đi
        start: Đỉnh bắt đầu
        target: Đỉnh đích
    """
    if not paths:
        print(f"Không tìm thấy đường đi từ {start} đến {target}")
        return
    
    print(f"Tìm thấy {len(paths)} đường đi từ {start} đến {target}:")
    print("-" * 50)
    
    for i, path in enumerate(paths, 1):
        # In từng đường đi với số thứ tự và độ dài
        path_str = " -> ".join(path)
        print(f"Đường đi {i}: {path_str} (Độ dài: {len(path) - 1} cạnh)")
    
    print("-" * 50)
    print(f"Tổng cộng: {len(paths)} đường đi")

# Định nghĩa đồ thị mẫu 7 (không trọng số, mật độ cao)
graph = {
    'S': ['A', 'D', 'E'],           # S kết nối với A, D, E
    'A': ['S', 'B', 'D', 'E'],      # A kết nối với S, B, D, E
    'B': ['A', 'C', 'E', 'F'],      # B kết nối với A, C, E, F
    'C': ['B', 'F', 'G'],           # C kết nối với B, F, G
    'D': ['S', 'A', 'E'],           # D kết nối với S, A, E
    'E': ['S', 'A', 'B', 'D', 'F', 'H'],  # E kết nối với S, A, B, D, F, H
    'F': ['B', 'C', 'E', 'G', 'H'], # F kết nối với B, C, E, G, H
    'G': ['C', 'F', 'H'],           # G kết nối với C, F, H
    'H': ['E', 'F', 'G']            # H kết nối với E, F, G
}

# Chạy thuật toán BFS để tìm tất cả đường đi từ S đến H
if __name__ == "__main__":
    start_node = 'S'
    target_node = 'H'
    
    print("=== THUẬT TOÁN BFS TÌM TẤT CẢ ĐƯỜNG ĐI ===")
    print(f"Đồ thị: {len(graph)} đỉnh")
    print(f"Tìm đường đi từ {start_node} đến {target_node}")
    print()
    
    # Tìm tất cả đường đi
    all_paths = bfs_all_paths(graph, start_node, target_node)
    
    # In kết quả
    print_all_paths(all_paths, start_node, target_node)
    
    # Thông tin bổ sung về đường đi ngắn nhất
    if all_paths:
        shortest_length = min(len(path) - 1 for path in all_paths)
        shortest_paths = [path for path in all_paths if len(path) - 1 == shortest_length]
        
        print(f"\nĐường đi ngắn nhất có độ dài: {shortest_length} cạnh")
        print(f"Số đường đi ngắn nhất: {len(shortest_paths)}")
        
        for i, path in enumerate(shortest_paths, 1):
            path_str = " -> ".join(path)
            print(f"  Đường {i}: {path_str}")
````

````python
# code here
# 3. Thêm một cạnh mới vào **Đồ thị mẫu 6** (ví dụ: B-H với trọng số 20). Chạy lại BFS và DFS, phân tích sự thay đổi trong đường đi và tổng trọng số.
from collections import deque, defaultdict

class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
        self.weights = {}
    
    def add_edge(self, u, v, weight):
        self.graph[u].append(v)
        self.weights[(u, v)] = weight
    
    def bfs(self, start, end):
        if start == end:
            return [start], 0
        
        visited = set()
        queue = deque([(start, [start], 0)])  # (node, path, total_weight)
        visited.add(start)
        
        while queue:
            node, path, total_weight = queue.popleft()
            
            for neighbor in self.graph[node]:
                if neighbor == end:
                    final_path = path + [neighbor]
                    final_weight = total_weight + self.weights[(node, neighbor)]
                    return final_path, final_weight
                
                if neighbor not in visited:
                    visited.add(neighbor)
                    new_path = path + [neighbor]
                    new_weight = total_weight + self.weights[(node, neighbor)]
                    queue.append((neighbor, new_path, new_weight))
        
        return None, 0
    
    def dfs(self, start, end, visited=None, path=None, weight=0):
        if visited is None:
            visited = set()
        if path is None:
            path = []
        
        visited.add(start)
        path = path + [start]
        
        if start == end:
            return path, weight
        
        for neighbor in self.graph[start]:
            if neighbor not in visited:
                edge_weight = self.weights[(start, neighbor)]
                result_path, result_weight = self.dfs(neighbor, end, visited.copy(), path, weight + edge_weight)
                if result_path:
                    return result_path, result_weight
        
        return None, 0

def analyze_graph(graph, name):
    print(f"\n=== {name} ===")
    bfs_path, bfs_weight = graph.bfs('S', 'H')
    dfs_path, dfs_weight = graph.dfs('S', 'H')
    
    print(f"BFS - Đường đi: {' -> '.join(bfs_path)}, Tổng trọng số: {bfs_weight}")
    print(f"DFS - Đường đi: {' -> '.join(dfs_path)}, Tổng trọng số: {dfs_weight}")
    
    return bfs_path, bfs_weight, dfs_path, dfs_weight

# Đồ thị mẫu 6 gốc
graph_original = Graph()
edges_original = [
    ('S', 'A', 2), ('S', 'C', 5), ('A', 'B', 3), ('A', 'D', 4),
    ('B', 'E', 6), ('C', 'D', 7), ('C', 'F', 9), ('D', 'E', 8),
    ('E', 'H', 10), ('F', 'G', 12), ('G', 'H', 15)
]

for u, v, w in edges_original:
    graph_original.add_edge(u, v, w)

# Đồ thị mẫu 6 với cạnh mới B-H (trọng số 20)
graph_modified = Graph()
edges_modified = edges_original + [('B', 'H', 20)]

for u, v, w in edges_modified:
    graph_modified.add_edge(u, v, w)

# Phân tích và so sánh
bfs1, bfs_w1, dfs1, dfs_w1 = analyze_graph(graph_original, "Đồ thị mẫu 6 gốc")
bfs2, bfs_w2, dfs2, dfs_w2 = analyze_graph(graph_modified, "Đồ thị mẫu 6 với cạnh B-H (trọng số 20)")

print("\n=== PHÂN TÍCH SỰ THAY ĐỔI ===")
print(f"BFS - Thay đổi đường đi: {'Có' if bfs1 != bfs2 else 'Không'}")
print(f"BFS - Thay đổi trọng số: {bfs_w2 - bfs_w1}")
print(f"DFS - Thay đổi đường đi: {'Có' if dfs1 != dfs2 else 'Không'}")
print(f"DFS - Thay đổi trọng số: {dfs_w2 - dfs_w1}")

print(f"\nCạnh B-H mới tạo đường đi trực tiếp từ B đến H với trọng số 20")
print(f"So với đường đi B->E->H (trọng số {6+10}=16), cạnh mới có trọng số cao hơn")
````

````python
# code here
# 1. Đo thời gian chạy của BFS và DFS trên **Đồ thị mẫu 6** và **Đồ thị mẫu 7** bằng thư viện `time`. So sánh hiệu suất và thêm chú thích vào mã.
import time
from collections import deque, defaultdict

# Đồ thị mẫu 6: Có trọng số, nhiều đường đi
graph6 = {
    'S': [('A', 2), ('C', 5)],
    'A': [('B', 3), ('D', 4)],
    'B': [('E', 6)],
    'C': [('D', 7), ('F', 9)],
    'D': [('E', 8)],
    'E': [('H', 10)],
    'F': [('G', 12)],
    'G': [('H', 15)],
    'H': []
}

# Đồ thị mẫu 7: Không trọng số, mật độ cao
graph7 = {
    'S': ['A', 'D', 'E'],
    'A': ['B', 'D', 'E'],
    'B': ['C', 'E', 'F'],
    'C': ['F', 'G'],
    'D': ['E'],
    'E': ['F', 'H'],
    'F': ['G', 'H'],
    'G': ['H'],
    'H': []
}

def bfs_weighted(graph, start, goal):
    """BFS cho đồ thị có trọng số (chỉ tìm đường đi, không tối ưu chi phí)"""
    queue = deque([start])
    visited = set([start])
    parent = {start: None}
    
    while queue:
        current = queue.popleft()
        if current == goal:
            # Xây dựng đường đi
            path = []
            while current:
                path.append(current)
                current = parent[current]
            return path[::-1]
        
        for neighbor, weight in graph[current]:
            if neighbor not in visited:
                visited.add(neighbor)
                parent[neighbor] = current
                queue.append(neighbor)
    
    return None

def dfs_weighted(graph, start, goal):
    """DFS cho đồ thị có trọng số"""
    stack = [start]
    visited = set()
    parent = {start: None}
    
    while stack:
        current = stack.pop()
        if current == goal:
            # Xây dựng đường đi
            path = []
            while current:
                path.append(current)
                current = parent[current]
            return path[::-1]
        
        if current not in visited:
            visited.add(current)
            for neighbor, weight in graph[current]:
                if neighbor not in visited:
                    parent[neighbor] = current
                    stack.append(neighbor)
    
    return None

def bfs_unweighted(graph, start, goal):
    """BFS cho đồ thị không trọng số"""
    queue = deque([start])
    visited = set([start])
    parent = {start: None}
    
    while queue:
        current = queue.popleft()
        if current == goal:
            # Xây dựng đường đi
            path = []
            while current:
                path.append(current)
                current = parent[current]
            return path[::-1]
        
        for neighbor in graph[current]:
            if neighbor not in visited:
                visited.add(neighbor)
                parent[neighbor] = current
                queue.append(neighbor)
    
    return None

def dfs_unweighted(graph, start, goal):
    """DFS cho đồ thị không trọng số"""
    stack = [start]
    visited = set()
    parent = {start: None}
    
    while stack:
        current = stack.pop()
        if current == goal:
            # Xây dựng đường đi
            path = []
            while current:
                path.append(current)
                current = parent[current]
            return path[::-1]
        
        if current not in visited:
            visited.add(current)
            for neighbor in graph[current]:
                if neighbor not in visited:
                    parent[neighbor] = current
                    stack.append(neighbor)
    
    return None

def benchmark_algorithm(func, *args, iterations=100000):
    """Đo thời gian chạy của thuật toán"""
    start_time = time.perf_counter()
    for _ in range(iterations):
        result = func(*args)
    end_time = time.perf_counter()
    return (end_time - start_time) / iterations, result

def safe_percentage_calculation(time1, time2):
    """Tính toán phần trăm an toàn, tránh chia cho 0"""
    if time2 == 0 or time1 == 0:
        return "không thể tính (thời gian quá nhỏ)"
    return f"{abs(time1/time2 - 1)*100:.1f}%"

# Test và so sánh hiệu suất
print("=== SO SÁNH HIỆU SUẤT BFS VÀ DFS ===\n")

print("1. Đồ thị mẫu 6 (Có trọng số, nhiều đường đi):")
print("   Đặc điểm: Có chu trình A-D-E-B-A, nhiều đường từ S đến H")

# Test đồ thị 6
bfs6_time, bfs6_path = benchmark_algorithm(bfs_weighted, graph6, 'S', 'H')
dfs6_time, dfs6_path = benchmark_algorithm(dfs_weighted, graph6, 'S', 'H')

print(f"   BFS: {bfs6_time:.8f}s - Đường đi: {' -> '.join(bfs6_path)}")
print(f"   DFS: {dfs6_time:.8f}s - Đường đi: {' -> '.join(dfs6_path)}")

ratio6 = safe_percentage_calculation(bfs6_time, dfs6_time)
faster6 = 'nhanh hơn' if dfs6_time < bfs6_time else 'chậm hơn'
print(f"   Tỷ lệ: DFS {faster6} BFS {ratio6}\n")

print("2. Đồ thị mẫu 7 (Không trọng số, mật độ cao):")
print("   Đặc điểm: Mật độ cạnh cao, nhiều chu trình, nhiều kết nối")

# Test đồ thị 7
bfs7_time, bfs7_path = benchmark_algorithm(bfs_unweighted, graph7, 'S', 'H')
dfs7_time, dfs7_path = benchmark_algorithm(dfs_unweighted, graph7, 'S', 'H')

print(f"   BFS: {bfs7_time:.8f}s - Đường đi: {' -> '.join(bfs7_path)}")
print(f"   DFS: {dfs7_time:.8f}s - Đường đi: {' -> '.join(dfs7_path)}")

ratio7 = safe_percentage_calculation(bfs7_time, dfs7_time)
faster7 = 'nhanh hơn' if dfs7_time < bfs7_time else 'chậm hơn'
print(f"   Tỷ lệ: DFS {faster7} BFS {ratio7}\n")

print("=== PHÂN TÍCH KẾT QUẢ ===")
print("• BFS (Breadth-First Search):")
print("  - Tìm đường đi ngắn nhất (ít bước nhất)")
print("  - Sử dụng queue (FIFO)")
print("  - Thích hợp cho đồ thị có nhiều nhánh")

print("\n• DFS (Depth-First Search):")
print("  - Tìm đường đi sâu trước")
print("  - Sử dụng stack (LIFO)")
print("  - Có thể nhanh hơn nếu đích gần nguồn theo chiều sâu")

print("\n• Nhận xét:")
print("  - Với đồ thị dày đặc: BFS thường ổn định hơn")
print("  - Với đồ thị thưa: DFS có thể nhanh hơn")
print("  - Hiệu suất phụ thuộc vào cấu trúc đồ thị và vị trí đích")
````

````python
# code here
# 2. Thiết kế một đồ thị có trọng số với ít nhất 10 nút và 15 cạnh. Chạy BFS và DFS, phân tích đường đi và tổng trọng số. Đảm bảo mã có chú thích chi tiết.
from collections import deque, defaultdict
import heapq

class WeightedGraph:
    def __init__(self):
        # Sử dụng dictionary để lưu trữ đồ thị
        # Key: nút, Value: list các tuple (nút_kề, trọng_số)
        self.graph = defaultdict(list)
        self.vertices = set()
    
    def add_edge(self, u, v, weight):
        """Thêm cạnh có trọng số vào đồ thị (vô hướng)"""
        self.graph[u].append((v, weight))
        self.graph[v].append((u, weight))
        self.vertices.add(u)
        self.vertices.add(v)
    
    def bfs(self, start, end=None):
        """
        Breadth-First Search (BFS)
        Trả về đường đi và tổng trọng số từ start đến end
        """
        if start not in self.vertices:
            return None, 0
        
        # Queue để lưu trữ các nút cần duyệt
        queue = deque([(start, [start], 0)])  # (nút_hiện_tại, đường_đi, tổng_trọng_số)
        visited = set([start])
        
        # Danh sách lưu tất cả đường đi từ start
        all_paths = []
        
        while queue:
            current, path, total_weight = queue.popleft()
            
            # Nếu tìm thấy đích (hoặc duyệt tất cả nếu end=None)
            if end and current == end:
                return path, total_weight
            
            # Duyệt tất cả nút kề
            for neighbor, weight in self.graph[current]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    new_path = path + [neighbor]
                    new_weight = total_weight + weight
                    queue.append((neighbor, new_path, new_weight))
                    
                    if not end:  # Nếu không chỉ định đích, lưu tất cả đường đi
                        all_paths.append((new_path, new_weight))
        
        return all_paths if not end else (None, 0)
    
    def dfs(self, start, end=None, visited=None, path=None, total_weight=0):
        """
        Depth-First Search (DFS) - Đệ quy
        Trả về đường đi và tổng trọng số từ start đến end
        """
        if visited is None:
            visited = set()
        if path is None:
            path = []
        
        visited.add(start)
        path.append(start)
        
        # Nếu tìm thấy đích
        if end and start == end:
            return path.copy(), total_weight
        
        # Danh sách lưu tất cả đường đi
        all_paths = []
        
        # Duyệt tất cả nút kề
        for neighbor, weight in self.graph[start]:
            if neighbor not in visited:
                if end:
                    result_path, result_weight = self.dfs(neighbor, end, visited.copy(), 
                                                        path.copy(), total_weight + weight)
                    if result_path:
                        return result_path, result_weight
                else:
                    # Đệ quy cho tất cả nút chưa thăm
                    sub_paths = self.dfs(neighbor, None, visited.copy(), 
                                       path.copy(), total_weight + weight)
                    if isinstance(sub_paths, list):
                        all_paths.extend(sub_paths)
                    else:
                        all_paths.append(sub_paths)
        
        if not end and len(path) > 1:
            all_paths.append((path.copy(), total_weight))
        
        return all_paths if not end else (None, 0)
    
    def dijkstra(self, start, end):
        """
        Thuật toán Dijkstra để tìm đường đi ngắn nhất
        """
        # Priority queue: (khoảng_cách, nút, đường_đi)
        pq = [(0, start, [start])]
        distances = {vertex: float('inf') for vertex in self.vertices}
        distances[start] = 0
        visited = set()
        
        while pq:
            current_dist, current_vertex, path = heapq.heappop(pq)
            
            if current_vertex in visited:
                continue
                
            visited.add(current_vertex)
            
            if current_vertex == end:
                return path, current_dist
            
            for neighbor, weight in self.graph[current_vertex]:
                distance = current_dist + weight
                
                if distance < distances[neighbor]:
                    distances[neighbor] = distance
                    heapq.heappush(pq, (distance, neighbor, path + [neighbor]))
        
        return None, float('inf')
    
    def display_graph(self):
        """Hiển thị cấu trúc đồ thị"""
        print("Cấu trúc đồ thị:")
        for vertex in sorted(self.vertices):
            connections = []
            for neighbor, weight in self.graph[vertex]:
                connections.append(f"{neighbor}(w:{weight})")
            print(f"Nút {vertex}: {' -> '.join(connections)}")

def main():
    # Tạo đồ thị có trọng số với 10 nút và 15 cạnh
    g = WeightedGraph()
    
    # Thêm 15 cạnh với trọng số khác nhau
    edges = [
        ('A', 'B', 4), ('A', 'C', 2), ('A', 'D', 7),
        ('B', 'E', 3), ('B', 'F', 5), ('C', 'F', 8),
        ('C', 'G', 6), ('D', 'G', 1), ('D', 'H', 9),
        ('E', 'I', 2), ('F', 'I', 4), ('F', 'J', 7),
        ('G', 'J', 3), ('H', 'J', 5), ('I', 'J', 1)
    ]
    
    print("=== THIẾT KẾ ĐỒ THỊ CÓ TRỌNG SỐ ===")
    print(f"Số nút: 10 (A, B, C, D, E, F, G, H, I, J)")
    print(f"Số cạnh: {len(edges)}")
    
    # Thêm các cạnh vào đồ thị
    for u, v, w in edges:
        g.add_edge(u, v, w)
    
    # Hiển thị cấu trúc đồ thị
    g.display_graph()
    
    print("\n" + "="*50)
    
    # Chạy BFS từ A đến J
    print("=== BREADTH-FIRST SEARCH (BFS) ===")
    start_node, end_node = 'A', 'J'
    bfs_path, bfs_weight = g.bfs(start_node, end_node)
    
    if bfs_path:
        print(f"BFS từ {start_node} đến {end_node}:")
        print(f"Đường đi: {' -> '.join(bfs_path)}")
        print(f"Tổng trọng số: {bfs_weight}")
    else:
        print(f"Không tìm thấy đường đi từ {start_node} đến {end_node}")
    
    print("\n" + "="*50)
    
    # Chạy DFS từ A đến J
    print("=== DEPTH-FIRST SEARCH (DFS) ===")
    dfs_path, dfs_weight = g.dfs(start_node, end_node)
    
    if dfs_path:
        print(f"DFS từ {start_node} đến {end_node}:")
        print(f"Đường đi: {' -> '.join(dfs_path)}")
        print(f"Tổng trọng số: {dfs_weight}")
    else:
        print(f"Không tìm thấy đường đi từ {start_node} đến {end_node}")
    
    print("\n" + "="*50)
    
    # Chạy Dijkstra để tìm đường đi ngắn nhất
    print("=== DIJKSTRA - ĐƯỜNG ĐI NGẮN NHẤT ===")
    dijkstra_path, dijkstra_weight = g.dijkstra(start_node, end_node)
    
    if dijkstra_path:
        print(f"Đường đi ngắn nhất từ {start_node} đến {end_node}:")
        print(f"Đường đi: {' -> '.join(dijkstra_path)}")
        print(f"Tổng trọng số: {dijkstra_weight}")
    
    print("\n" + "="*50)
    
    # Phân tích so sánh
    print("=== PHÂN TÍCH SO SÁNH ===")
    print(f"BFS - Tổng trọng số: {bfs_weight}")
    print(f"DFS - Tổng trọng số: {dfs_weight}")
    print(f"Dijkstra - Tổng trọng số: {dijkstra_weight}")
    
    print("\nNhận xét:")
    print("- BFS tìm đường đi với ít nút nhất, nhưng không đảm bảo trọng số nhỏ nhất")
    print("- DFS có thể tìm được đường đi khác với BFS, phụ thuộc vào thứ tự duyệt")
    print("- Dijkstra luôn tìm được đường đi có tổng trọng số nhỏ nhất")
    
    # Hiển thị một số đường đi khác từ A
    print(f"\n=== MỘT SỐ ĐƯỜNG ĐI KHÁC TỪ {start_node} ===")
    other_nodes = ['E', 'G', 'H']
    for node in other_nodes:
        path, weight = g.dijkstra(start_node, node)
        if path:
            print(f"Đến {node}: {' -> '.join(path)} (Trọng số: {weight})")

if __name__ == "__main__":
    main()
````
