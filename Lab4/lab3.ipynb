## Lý thuyết
Thuật toán di truyền (Genetic Algorithm - GA) là một phương pháp tìm kiếm dựa trên cơ chế tiến hóa tự nhiên, lấy cảm hứng từ thuyết tiến hóa của Darwin. GA được sử dụng để tìm lời giải gần tối ưu cho các bài toán tối ưu hóa phức tạp.

### Các bước chính của thuật toán:
1. **Khởi tạo quần thể**: Tạo ngẫu nhiên một tập hợp các cá thể (giải pháp tiềm năng).
2. **Đánh giá độ thích nghi (Fitness Evaluation)**: Tính giá trị hàm mục tiêu cho mỗi cá thể.
3. **Lựa chọn (Selection)**: Chọn các cá thể tốt nhất để lai ghép (thường sử dụng các phương pháp như Tournament Selection hoặc Roulette Wheel).
4. **Lai ghép (Crossover)**: Kết hợp hai cá thể để tạo ra cá thể mới.
5. **Đột biến (Mutation)**: Thay đổi ngẫu nhiên một số gen để duy trì tính đa dạng.
6. **Lặp lại**: Quay lại bước 2 cho đến khi đạt điều kiện dừng (số thế hệ hoặc giá trị fitness mong muốn).

## Ví dụ 1: Tối ưu hóa hàm một biến

### Bài toán  
Tìm $x$ sao cho  
$$
f(x) = -(x^2) + 10x + 50
$$  
đạt giá trị lớn nhất trong khoảng  
$$
x \in [0, 10].
$$

# Mã giả
```plaintext
Tạo quần thể mới rỗng
Lặp cho đến khi quần thể mới đủ N cá thể:
    Chọn cha mẹ 1 bằng Tournament Selection:
        - Chọn ngẫu nhiên tournament_size cá thể trong quần thể
        - Chọn cá thể có fitness cao nhất làm cha mẹ 1
    Chọn cha mẹ 2 bằng Tournament Selection tương tự
    
    Thực hiện lai ghép (crossover) giữa cha mẹ 1 và cha mẹ 2 với xác suất crossover_rate:
        Nếu crossover thành công:
            Tạo con mới có giá trị trung bình của cha mẹ
        Ngược lại:
            Con mới là cha mẹ 1
            
    Thực hiện đột biến cho con mới với xác suất mutation_rate:
        Nếu đột biến:
            Gán con mới giá trị ngẫu nhiên trong [min_val, max_val]
            
    Thêm con mới vào quần thể mới
    
Cập nhật quần thể hiện tại = quần thể mới
In ra cá thể tốt nhất và giá trị fitness tương ứng của thế hệ


````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Hàm mục tiêu
def fitness_function(x):
    return -(x**2) + 10*x + 50

# Khởi tạo quần thể
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# Lựa chọn (Tournament Selection)
def select_parents(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghép
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# Đột biến
def mutate(individual, mutation_rate=0.1, min_val=0, max_val=10):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thuật toán di truyền
def genetic_algorithm_example1(pop_size=50, generations=100, min_val=0, max_val=10):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function(x) for x in population])]
        print(f'Thế hệ {generation + 1}: x = {best_individual:.4f}, f(x) = {best_fitness:.4f}')
    
    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]
    
    # Trực quan hóa
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị f(x) tốt nhất')
    plt.title('Sự thay đổi của f(x) qua các thế hệ')
    plt.show()
    
    return best_individual, best_fitness

# Chạy thuật toán
best_x, best_f = genetic_algorithm_example1()
print(f'\nKết quả cuối cùng: x = {best_x:.4f}, f(x) = {best_f:.4f}')
````

```plaintext
Tạo quần thể mới rỗng
Lặp cho đến khi quần thể mới đủ N cá thể:
    Chọn cha mẹ 1 bằng Tournament Selection trên quần thể hiện tại
    Chọn cha mẹ 2 bằng Tournament Selection trên quần thể hiện tại
    Thực hiện lai ghép (crossover) giữa cha mẹ 1 và cha mẹ 2 với xác suất crossover_rate:
        Nếu crossover thành công:
            Tạo con mới bằng trung bình giá trị (x, y) của cha mẹ
        Ngược lại:
            Con mới là cha mẹ 1
    Thực hiện đột biến cho con mới với xác suất mutation_rate từng biến:

In kết quả tốt nhất thế hệ

````python
# Hàm mục tiêu
def fitness_function_2d(x, y):
    return -(x**2 + y**2)  # Sử dụng dấu âm vì GA tìm max

# Khởi tạo quần thể
def initialize_population_2d(pop_size, min_val, max_val):
    return [(random.uniform(min_val, max_val), random.uniform(min_val, max_val)) for _ in range(pop_size)]

# Lựa chọn
def select_parents_2d(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghép
def crossover_2d(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return ((parent1[0] + parent2[0]) / 2, (parent1[1] + parent2[1]) / 2)
    return parent1

# Đột biến
def mutate_2d(individual, mutation_rate=0.1, min_val=-5, max_val=5):
    x, y = individual
    if random.random() < mutation_rate:
        x = random.uniform(min_val, max_val)
    if random.random() < mutation_rate:
        y = random.uniform(min_val, max_val)
    return (x, y)

# Thuật toán di truyền
def genetic_algorithm_example2(pop_size=50, generations=100, min_val=-5, max_val=5):
    population = initialize_population_2d(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function_2d(x, y) for x, y in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(-best_fitness)  # Chuyển về giá trị thực
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents_2d(population, fitness_values)
            parent2 = select_parents_2d(population, fitness_values)
            child = crossover_2d(parent1, parent2)
            child = mutate_2d(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
        print(f'Thế hệ {generation + 1}: (x, y) = ({best_individual[0]:.4f}, {best_individual[1]:.4f}), g(x, y) = {-best_fitness:.4f}')
    
    best_fitness = max([fitness_function_2d(x, y) for x, y in population])
    best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
    
    # Trực quan hóa
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị g(x, y) tốt nhất')
    plt.title('Sự thay đổi của g(x, y) qua các thế hệ')
    plt.show()
    
    return best_individual, -best_fitness

# Chạy thuật toán
best_xy, best_g = genetic_algorithm_example2()
print(f'\nKết quả cuối cùng: (x, y) = ({best_xy[0]:.4f}, {best_xy[1]:.4f}), g(x, y) = {best_g:.4f}')
````

# Bài tập 1
# Thay đổi hàm mục tiêu**: Thử áp dụng thuật toán cho hàm $h(x) = sin(x) + cos(x)$ trong khoảng $x \in [0, 2\pi]$. So sánh kết quả với nghiệm chính xác.

## Tối ưu hàm $h(x) = \sin(x) + \cos(x)$ bằng Thuật Toán Di Truyền (Genetic Algorithm)

## Mục tiêu
- Tối ưu hàm:
  $$
  h(x) = \sin(x) + \cos(x)
  $$
  trong miền:
  $$
  x \in [0, 2\pi]
  $$
- Tìm giá trị **x** sao cho **h(x)** đạt giá trị lớn nhất.

---

## Phân tích nghiệm chính xác

Ta có:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xảy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

## Ý tưởng thuật toán di truyền (GA)

1. **Mã hóa**: Biểu diễn giá trị \( x \) bằng chuỗi nhị phân dài 16 bit.  
2. **Giải mã**: Chuyển từ chuỗi nhị phân sang giá trị thực trong khoảng \([0, 2\pi]\).  
3. **Fitness**: Sử dụng chính giá trị \( h(x) \) làm độ thích nghi.  
4. **Chọn lọc**: Roulette Wheel Selection.  
5. **Lai ghép**: Lai 1 điểm với xác suất cao.  
6. **Đột biến**: Lật bit với xác suất thấp.

---


````python
# code vào  đây nhe
# Giải Bài tập 1 bằng Thuật toán Di truyền (Genetic Algorithm)
# -----------------------------------------------------------
# Mục tiêu: Tối đa hóa h(x) = sin(x) + cos(x) trên [0, 2π]
# - Mã hóa nhị phân 16‑bit
# - Chọn lọc Roulette Wheel
# - Lai ghép 1 điểm
# - Đột biến lật bit
# -----------------------------------------------------------
import math
import random
import matplotlib.pyplot as plt

# ----------------------- Tham số GA ------------------------
BITS = 16               # số bit mã hóa
POP_SIZE = 50           # kích thước quần thể
GENERATIONS = 100       # số thế hệ
CROSS_RATE = 0.9        # xác suất lai
MUT_RATE = 0.005        # xác suất đột biến mỗi bit
LOW, HIGH = 0.0, 2 * math.pi  # miền tìm kiếm

# ------------- Hàm mục tiêu & trợ giúp --------------------
def h(x: float) -> float:
    """Hàm mục tiêu cần tối đa hóa."""
    return math.sin(x) + math.cos(x)

def encode(x: float) -> int:
    """Chuyển giá trị thực x -> gene nhị phân (kiểu int)."""
    return int(round((x - LOW) / (HIGH - LOW) * (2**BITS - 1)))

def decode(gene: int) -> float:
    """Chuyển gene nhị phân -> giá trị thực x."""
    return LOW + gene / (2**BITS - 1) * (HIGH - LOW)

def random_gene() -> int:
    """Sinh gene ngẫu nhiên."""
    return random.randint(0, 2**BITS - 1)

def gene_fitness(gene: int) -> float:
    """Độ thích nghi = giá trị hàm h(x)."""
    return h(decode(gene))

# -------------------- Các toán tử GA ----------------------
def roulette_select(pop):
    """Chọn lọc Roulette Wheel – tỷ lệ thuận fitness (đã dịch >0)."""
    fits = [gene_fitness(g) for g in pop]
    min_fit = min(fits)
    shifted = [f - min_fit + 1e-9 for f in fits]  # đảm bảo >0
    total = sum(shifted)
    r = random.uniform(0, total)
    acc = 0.0
    for g, w in zip(pop, shifted):
        acc += w
        if acc >= r:
            return g
    return pop[-1]

def crossover(g1: int, g2: int):
    """Lai ghép 1 điểm, trả về 2 con."""
    if random.random() < CROSS_RATE:
        point = random.randint(1, BITS - 1)
        mask = (1 << point) - 1
        child1 = (g1 & ~mask) | (g2 & mask)
        child2 = (g2 & ~mask) | (g1 & mask)
        return child1, child2
    return g1, g2

def mutate(g: int) -> int:
    """Đột biến: lật ngẫu nhiên các bit với xác suất MUT_RATE."""
    for bit in range(BITS):
        if random.random() < MUT_RATE:
            g ^= (1 << bit)
    return g

# ----------------------- Chạy GA ---------------------------
random.seed(42)  # cố định seed để tái lập kết quả
population = [random_gene() for _ in range(POP_SIZE)]
best_history = []  # lưu giá trị tốt nhất qua các thế hệ

for gen in range(1, GENERATIONS + 1):
    best_gene = max(population, key=gene_fitness)   # cá thể tốt nhất
    best_x = decode(best_gene)
    best_h = h(best_x)
    
    # In log theo định dạng yêu cầu
    print(f"Thế hệ {gen}: x = {best_x: .4f}, h(x) = {best_h: .4f}")
    
    best_history.append(best_h)
    
    # Tạo quần thể mới
    new_pop = []
    while len(new_pop) < POP_SIZE:
        p1, p2 = roulette_select(population), roulette_select(population)
        c1, c2 = crossover(p1, p2)
        new_pop.extend([mutate(c1), mutate(c2)])
    population = new_pop[:POP_SIZE]

# ----------------- So sánh với nghiệm chính xác ------------
exact_x = math.pi / 4
exact_h = math.sqrt(2)
print("\n--- So sánh với nghiệm chính xác ---")
print(f"GA:     x* = {best_x:.6f} rad ({math.degrees(best_x):.4f}°), h(x*) = {best_h:.6f}")
print(f"Exact:  x* = {exact_x:.6f} rad ({math.degrees(exact_x):.4f}°), h(x*) = {exact_h:.6f}")
print(f"Δx = {abs(best_x - exact_x):.6e} rad, Δh = {abs(best_h - exact_h):.6e}")

# -------------------- Vẽ biểu đồ hội tụ --------------------
plt.figure(figsize=(8, 4))
plt.plot(range(1, GENERATIONS + 1), best_history, linewidth=2)
plt.title("Biểu đồ hội tụ GA – h(x) tốt nhất theo thế hệ")
plt.xlabel("Thế hệ")
plt.ylabel("Giá trị h(x) tốt nhất")
plt.grid(True)
plt.tight_layout()
plt.show()
````

# Bài tập 2: Tối ưu hàm $h(x) = \sin(x) + \cos(x)$ bằng Thuật Toán Di Truyền (Genetic Algorithm)

---

## Mục tiêu

Tối ưu hàm:
$$
h(x) = \sin(x) + \cos(x)
$$
trong khoảng:
$$
x \in [0, 2\pi]
$$

> Tìm giá trị $x$ sao cho $h(x)$ đạt **giá trị lớn nhất**.

---

## Phân tích nghiệm chính xác

Ta có:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xảy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##  Ý tưởng thuật toán di truyền (GA)

| Thành phần        | Chi tiết                                                                 |
|------------------|-------------------------------------------------------------------------|
| **Mã hóa**        | Dùng chuỗi nhị phân 16 bit để biểu diễn giá trị $x \in [0, 2\pi]$         |
| **Giải mã**       | Ánh xạ nhị phân sang giá trị thực: $x = 0 + \dfrac{\text{decimal}}{2^{16}-1} \cdot 2\pi$ |
| **Fitness**       | Chính là $h(x) = \sin(x) + \cos(x)$                                     |
| **Chọn lọc**      | Roulette Wheel Selection                                                |
| **Lai ghép**      | 1 điểm cắt với xác suất 0.8                                              |
| **Đột biến**      | Xác suất 5% (0.05)                                                       |

---

````python
# code vào đây nhe
# Bài tập 2 – Tối ưu h(x)=sin(x)+cos(x) bằng GA
# ------------------------------------------------
# Tham số theo đề: 16 bit, roulette, crossover 1‑điểm (p=0.8), mutation 5%
# ------------------------------------------------
import math, random, matplotlib.pyplot as plt

# ---------------- Tham số GA --------------------
BITS       = 16          # số bit mã hóa
POP_SIZE   = 50          # kích thước quần thể
GENERATIONS= 100         # số thế hệ
CROSS_RATE = 0.8         # xác suất lai
MUT_RATE   = 0.05        # xác suất đột biến mỗi bit
LOW, HIGH  = 0.0, 2*math.pi

# --------------- Hàm mục tiêu -------------------
def h(x: float) -> float:
    return math.sin(x) + math.cos(x)

# -------------- Mã hóa / giải mã ----------------
def encode(x: float) -> int:
    return int(round((x - LOW) / (HIGH - LOW) * (2**BITS - 1)))

def decode(gene: int) -> float:
    return LOW + gene / (2**BITS - 1) * (HIGH - LOW)

# -------------- Độ thích nghi -------------------
def fitness(gene: int) -> float:
    return h(decode(gene))

# -------------- Chọn lọc roulette ---------------
def roulette_select(pop):
    fits = [fitness(g) for g in pop]
    min_fit = min(fits)
    shifted = [f - min_fit + 1e-9 for f in fits]  # dịch về >0
    s = sum(shifted)
    r = random.uniform(0, s)
    acc = 0.0
    for g, w in zip(pop, shifted):
        acc += w
        if acc >= r:
            return g
    return pop[-1]

# ---------------- Lai – 1 điểm ------------------
def crossover(g1: int, g2: int):
    if random.random() < CROSS_RATE:
        point = random.randint(1, BITS - 1)
        mask = (1 << point) - 1
        return (g1 & ~mask) | (g2 & mask), (g2 & ~mask) | (g1 & mask)
    return g1, g2

# ----------------- Đột biến ---------------------
def mutate(g: int) -> int:
    for b in range(BITS):
        if random.random() < MUT_RATE:
            g ^= (1 << b)   # lật bit b
    return g

# --------------- Chạy GA ------------------------
random.seed(2025)
pop = [random.randint(0, 2**BITS - 1) for _ in range(POP_SIZE)]
best_history = []

for gen in range(1, GENERATIONS + 1):
    best_gene = max(pop, key=fitness)
    best_x = decode(best_gene)
    best_h = h(best_x)
    print(f"Thế hệ {gen:3d}: x = {best_x: .4f}, h(x) = {best_h: .4f}")
    best_history.append(best_h)
    
    # sinh thế hệ mới
    new_pop = []
    while len(new_pop) < POP_SIZE:
        p1, p2 = roulette_select(pop), roulette_select(pop)
        c1, c2 = crossover(p1, p2)
        new_pop.extend([mutate(c1), mutate(c2)])
    pop = new_pop[:POP_SIZE]

# -------- So sánh với nghiệm chính xác ---------
exact_x, exact_h = math.pi/4, math.sqrt(2)
print("\n==> Nghiệm GA:   x* = %.6f rad (%.4f°), h(x*) = %.6f" % (best_x, math.degrees(best_x), best_h))
print("    Nghiệm chuẩn: x* = %.6f rad (%.4f°), h(x*) = %.6f" % (exact_x, math.degrees(exact_x), exact_h))
print("    Δx = %.3e rad, Δh = %.3e" % (abs(best_x-exact_x), abs(best_h-exact_h)))

# --------------- Biểu đồ hội tụ -----------------
plt.figure(figsize=(8,4))
plt.plot(range(1, GENERATIONS+1), best_history, lw=2)
plt.title("Biểu đồ hội tụ GA – h(x) tốt nhất qua các thế hệ")
plt.xlabel("Thế hệ")
plt.ylabel("h(x) tốt nhất")
plt.grid(True)
plt.tight_layout()
plt.show()
````

# Thay đổi tham số thuật toán di truyền và phân tích tốc độ hội tụ

**Mục tiêu:** Thử thay đổi các tham số chính của thuật toán di truyền gồm:

- Kích thước quần thể: `pop_size`
- Số thế hệ: `generations`
- Tỷ lệ đột biến: `mutation_rate`
- Tỷ lệ lai ghép: `crossover_rate`

và phân tích ảnh hưởng của từng tham số đến tốc độ hội tụ của thuật toán.

---

## Các tham số và vai trò

| Tham số         | Ý nghĩa                                  | Ảnh hưởng chính                                                                                         |
|-----------------|-----------------------------------------|--------------------------------------------------------------------------------------------------------|
| `pop_size`      | Số cá thể trong quần thể                 | Quần thể lớn giúp khám phá không gian tìm kiếm tốt hơn nhưng tốn nhiều thời gian tính toán. Quần thể nhỏ hội tụ nhanh nhưng dễ bị kẹt cực trị cục bộ. |
| `generations`   | Số vòng lặp của thuật toán               | Số thế hệ càng nhiều thì thuật toán có nhiều cơ hội cải thiện nghiệm, tuy nhiên thời gian chạy cũng tăng. |
| `mutation_rate` | Xác suất đột biến mỗi bit                | Đột biến cao giúp tăng đa dạng gene, tránh bị kẹt trong cực trị cục bộ, nhưng nếu quá cao làm quá trình tìm kiếm nhiễu loạn. |
| `crossover_rate`| Xác suất lai ghép giữa hai cá thể       | Lai ghép giúp phối hợp gen tốt, tăng khả năng tìm nghiệm tốt. Nếu quá thấp, quần thể kém đa dạng; quá cao có thể phá hủy cấu trúc tốt của cá thể. |

---

## Phân tích ảnh hưởng

- **`pop_size`**: Kích thước quần thể lớn giúp tìm kiếm toàn cục tốt hơn nhưng chậm hơn, kích thước nhỏ làm thuật toán hội tụ nhanh nhưng dễ mắc kẹt.
- **`generations`**: Số thế hệ lớn tăng khả năng cải tiến kết quả nhưng làm tăng thời gian tính toán.
- **`mutation_rate`**: Tỷ lệ đột biến vừa phải giúp duy trì đa dạng di truyền, tránh kẹt cục bộ; quá cao hoặc quá thấp đều làm giảm hiệu quả.
- **`crossover_rate`**: Tỷ lệ lai ghép hợp lý giúp khai thác tốt thông tin di truyền, tăng hiệu quả hội tụ.

---




````python
# code vào đây 
"""
GA Parameter Study – h(x) = sin(x) + cos(x)
- Mã hóa 16-bit, roulette, crossover 1 điểm, mutation lật bit
- Thử nhiều giá trị pop_size, generations, mutation_rate, crossover_rate
- Đánh giá: thế hệ đầu đạt ngưỡng (≈√2 - 1e-4) & h(x) cuối
- Hiển thị bảng kết quả (pandas) + biểu đồ Matplotlib
"""
import math, random, pandas as pd, matplotlib.pyplot as plt
from typing import List, Dict

# ---------- Cấu hình & hằng số ----------
BITS       = 16
LOW, HIGH  = 0.0, 2*math.pi
DEFAULTS   = dict(pop_size=50, generations=100,
                  crossover_rate=0.8, mutation_rate=0.05)
THRESHOLD  = math.sqrt(2) - 1e-4  # ~1.414114

# ---------- Hàm mục tiêu & GA ----------
def h(x: float) -> float:
    return math.sin(x) + math.cos(x)

def encode(x: float) -> int:
    return int(round((x-LOW)/(HIGH-LOW)*(2**BITS-1)))

def decode(g: int) -> float:
    return LOW + g/(2**BITS-1)*(HIGH-LOW)

def fitness(g: int) -> float:
    return h(decode(g))

def roulette(pop):
    fits = [fitness(g) for g in pop]
    m = min(fits)
    shifted = [f-m+1e-9 for f in fits]
    total = sum(shifted)
    r = random.uniform(0, total)
    acc = 0
    for g,w in zip(pop, shifted):
        acc += w
        if acc >= r:
            return g
    return pop[-1]

def crossover(g1, g2, rate):
    if random.random() < rate:
        p = random.randint(1, BITS-1)
        mask = (1<<p)-1
        return (g1&~mask)|(g2&mask), (g2&~mask)|(g1&mask)
    return g1, g2

def mutate(g, rate):
    for b in range(BITS):
        if random.random() < rate:
            g ^= (1<<b)
    return g

def run_ga(pop_size, generations, crossover_rate, mutation_rate, seed=0)->List[float]:
    random.seed(seed)
    pop = [random.randint(0, 2**BITS-1) for _ in range(pop_size)]
    best_hist=[]
    for _ in range(generations):
        best_gene = max(pop, key=fitness)
        best_hist.append(fitness(best_gene))
        new=[]
        while len(new) < pop_size:
            p1,p2 = roulette(pop), roulette(pop)
            c1,c2 = crossover(p1,p2,crossover_rate)
            new.extend([mutate(c1,mutation_rate), mutate(c2,mutation_rate)])
        pop = new[:pop_size]
    return best_hist

def first_hit(hist, thr=THRESHOLD):
    for i,v in enumerate(hist,1):
        if v >= thr:
            return i
    return None

# ---------- Bộ cấu hình thử nghiệm ----------
tests=[]
for ps in [20,50,100]:
    cfg=DEFAULTS.copy(); cfg['pop_size']=ps
    tests.append(("pop_size",ps,cfg))
for g in [30,60,100]:
    cfg=DEFAULTS.copy(); cfg['generations']=g
    tests.append(("generations",g,cfg))
for mr in [0.01,0.05,0.1]:
    cfg=DEFAULTS.copy(); cfg['mutation_rate']=mr
    tests.append(("mutation_rate",mr,cfg))
for cr in [0.6,0.8,0.95]:
    cfg=DEFAULTS.copy(); cfg['crossover_rate']=cr
    tests.append(("crossover_rate",cr,cfg))

# ---------- Chạy & thu kết quả ----------
records=[]
histories: Dict[str,Dict[float,List[float]]] = {}
for param,val,cfg in tests:
    hist = run_ga(**cfg, seed=2025)
    histories.setdefault(param,{})[val]=hist
    records.append(dict(Parameter=param, Value=val,
                        Gen_hit=first_hit(hist),
                        Best_final=hist[-1]))

df = pd.DataFrame(records)
print("\nBẢNG KẾT QUẢ:")
print(df.to_string(index=False))

# ---------- Hàm vẽ ----------
def plot_group(name, dic):
    plt.figure(figsize=(8,4))
    for val,hist in dic.items():
        plt.plot(range(1,len(hist)+1), hist, label=str(val))
    plt.title(f"Ảnh hưởng {name}")
    plt.xlabel("Thế hệ"); plt.ylabel("h(x) tốt nhất")
    plt.grid(True); plt.legend(); plt.tight_layout(); plt.show()

for group in histories:
    plot_group(group, histories[group])
````

````python
# code vào đây nhé
"""
BÀI TẬP 4 – So sánh ba phương pháp chọn lọc GA
==============================================

• Tối ưu h(x)=sin(x)+cos(x) trên [0,2π]  
• Ba phương pháp chọn lọc: Random, Roulette, Tournament  
• Tham số cố định: 16 bit, pop=50, gen=100, crossover=0.8, mutation=0.05  
• Chạy 20 seed → tính trung bình:  
  – Gen đầu tiên đạt ngưỡng (≈√2 − 1e‑4)  
  – Giá trị h(x) cuối  
• Vẽ biểu đồ fitness trung bình theo thế hệ
"""
import math, random, statistics, pandas as pd, matplotlib.pyplot as plt
from typing import List, Dict, Callable

# ------------------ Cấu hình chung ------------------
BITS = 16
LOW, HIGH = 0.0, 2*math.pi
POP_SIZE = 50
GENERATIONS = 100
CROSS_RATE = 0.8
MUT_RATE = 0.05
THRESHOLD = math.sqrt(2) - 1e-4  # ngưỡng hội tụ
RUNS = 20                         # số lần chạy mỗi phương pháp

# ------------------ Hàm GA cơ bản -------------------
def h(x): return math.sin(x) + math.cos(x)
def encode(x): return int(round((x-LOW)/(HIGH-LOW)*(2**BITS-1)))
def decode(g): return LOW + g/(2**BITS-1)*(HIGH-LOW)
def fitness(g): return h(decode(g))

# --- Chọn lọc ---
def sel_random(pop): return random.choice(pop)

def sel_roulette(pop):
    fits = [fitness(g) for g in pop]
    m = min(fits)
    shifted = [f - m + 1e-9 for f in fits]
    total = sum(shifted)
    pick = random.uniform(0,total)
    acc=0
    for g,w in zip(pop,shifted):
        acc += w
        if acc >= pick:
            return g
    return pop[-1]

def sel_tournament(pop, k=3):
    return max(random.sample(pop,k), key=fitness)

# --- Toán tử di truyền ---
def crossover(g1,g2):
    if random.random()<CROSS_RATE:
        p=random.randint(1,BITS-1)
        mask=(1<<p)-1
        return (g1&~mask)|(g2&mask), (g2&~mask)|(g1&mask)
    return g1,g2
def mutate(g):
    for b in range(BITS):
        if random.random()<MUT_RATE:
            g ^= (1<<b)
    return g

# --- Hàm chạy GA một lần ---
def run_ga(selector:Callable[[List[int]],int], seed:int)->List[float]:
    random.seed(seed)
    pop=[random.randint(0,2**BITS-1) for _ in range(POP_SIZE)]
    best_hist=[]
    for _ in range(GENERATIONS):
        best_hist.append(fitness(max(pop,key=fitness)))
        new=[]
        while len(new)<POP_SIZE:
            p1,p2=selector(pop),selector(pop)
            c1,c2=crossover(p1,p2)
            new.extend([mutate(c1), mutate(c2)])
        pop=new[:POP_SIZE]
    return best_hist

def first_hit(hist, thr=THRESHOLD):
    for i,v in enumerate(hist,1):
        if v>=thr: return i
    return None

methods = {
    "Random":    sel_random,
    "Roulette":  sel_roulette,
    "Tournament":lambda pop: sel_tournament(pop,3)
}

records=[]
avg_histories={}  # mean fitness per gen
for name,selector in methods.items():
    all_runs=[]
    gen_hits=[]
    best_finals=[]
    for seed in range(RUNS):
        hist=run_ga(selector, seed)
        all_runs.append(hist)
        gen_hits.append(first_hit(hist) or GENERATIONS)
        best_finals.append(hist[-1])
    # trung bình theo thế hệ
    avg_hist=[statistics.mean(col) for col in zip(*all_runs)]
    avg_histories[name]=avg_hist
    records.append({
        "Method":name,
        "Avg Gen Hit": statistics.mean(gen_hits),
        "Std Gen Hit": statistics.stdev(gen_hits),
        "Avg Final h(x)": statistics.mean(best_finals)
    })

# ---------- Bảng kết quả ----------
df=pd.DataFrame(records)
print("\nKẾT QUẢ TRUNG BÌNH SAU", RUNS, "LẦN CHẠY:")
print(df.to_string(index=False))

# ---------- Biểu đồ hội tụ ----------
plt.figure(figsize=(8,4))
for name,hist in avg_histories.items():
    plt.plot(range(1,GENERATIONS+1), hist, label=name)
plt.title("Fitness trung bình theo thế hệ (20 lần chạy)")
plt.xlabel("Thế hệ")
plt.ylabel("h(x) tốt nhất (trung bình)")
plt.grid(True); plt.legend(); plt.tight_layout(); plt.show()
````

# Bài tập 5
# Trực quan hóa quần thể – Vẽ biểu đồ phân tán (scatter plot) thể hiện sự phân bố của quần thể qua các thế hệ

---

##  Mục tiêu

- Trực quan hóa sự phân bố của quần thể cá thể trong quá trình tiến hóa của thuật toán di truyền.
- Dùng biểu đồ phân tán (scatter plot) để thể hiện vị trí các cá thể trên không gian biến số qua từng thế hệ.
- Giúp quan sát rõ tiến trình hội tụ, đa dạng di truyền, và sự thay đổi quần thể theo thời gian.

---

## Ý tưởng thực hiện

- Với bài toán tối ưu hàm $k(x, y, z) = x^2 + y^2 + z^2$, không gian tìm kiếm là 3 chiều.
- Mỗi cá thể là một điểm trong không gian $(x, y, z)$.
- Ở mỗi thế hệ, lấy vị trí giải mã của toàn bộ cá thể, vẽ scatter plot.
- Nếu không thể trực quan 3D trực tiếp, có thể vẽ scatter 2D từng cặp biến: $(x, y)$, $(y, z)$, $(x, z)$ qua các thế hệ.

---

````python
#code vào đây
import math, random, matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ------------------- GA parameters -------------------
POP_SIZE = 80
GENERATIONS = 100
CROSS_RATE = 0.8
MUT_RATE = 0.1
LOW, HIGH = -5.0, 5.0  # search space for each dimension (x,y,z)

# ------------------- Objective function --------------
def k(x, y, z):
    """k(x,y,z) = x^2 + y^2 + z^2 (minimize -> optimum at (0,0,0))."""
    return x**2 + y**2 + z**2

# ---------------- Individual helpers -----------------
def random_individual():
    return [random.uniform(LOW, HIGH) for _ in range(3)]

def fitness(ind):
    # negate because we want to maximise in GA framework
    return -k(*ind)

def mutate(ind):
    for i in range(3):
        if random.random() < MUT_RATE:
            ind[i] += random.gauss(0, 0.5)
            ind[i] = max(LOW, min(HIGH, ind[i]))
    return ind

def crossover(p1, p2):
    if random.random() < CROSS_RATE:
        alpha = random.random()
        return [
            alpha * p1[i] + (1 - alpha) * p2[i] for i in range(3)
        ], [
            (1 - alpha) * p1[i] + alpha * p2[i] for i in range(3)
        ]
    return p1, p2

def tournament_select(pop, k=3):
    return max(random.sample(pop, k), key=fitness)

# ------------------- GA run with logging -------------
random.seed(1)
population = [random_individual() for _ in range(POP_SIZE)]
pop_history = [population.copy()]

for gen in range(1, GENERATIONS + 1):
    new_population = []
    while len(new_population) < POP_SIZE:
        p1 = tournament_select(population)
        p2 = tournament_select(population)
        c1, c2 = crossover(p1, p2)
        new_population.extend([mutate(c1), mutate(c2)])
    population = new_population[:POP_SIZE]
    pop_history.append(population.copy())

# --------------- Visualization (scatter) -------------
def plot_generation(pop, gen_idx):
    xs = [p[0] for p in pop]
    ys = [p[1] for p in pop]
    zs = [p[2] for p in pop]
    
    fig = plt.figure(figsize=(6,5))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(xs, ys, zs)
    ax.set_title(f"Phân bố quần thể – Thế hệ {gen_idx}")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")
    plt.tight_layout()
    plt.show()

# Plot for thế hệ 0, 10, 50, 100
for idx in [0, 10, 50, 100]:
    plot_generation(pop_history[idx], idx)
````
