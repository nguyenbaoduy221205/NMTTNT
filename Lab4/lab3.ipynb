## L thuyt
Thut ton di truyn (Genetic Algorithm - GA) l mt phng php tm kim da trn c ch tin ha t nhin, ly cm hng t thuyt tin ha ca Darwin. GA c s dng  tm li gii gn ti u cho cc bi ton ti u ha phc tp.

### Cc bc chnh ca thut ton:
1. **Khi to qun th**: To ngu nhin mt tp hp cc c th (gii php tim nng).
2. **nh gi  thch nghi (Fitness Evaluation)**: Tnh gi tr hm mc tiu cho mi c th.
3. **La chn (Selection)**: Chn cc c th tt nht  lai ghp (thng s dng cc phng php nh Tournament Selection hoc Roulette Wheel).
4. **Lai ghp (Crossover)**: Kt hp hai c th  to ra c th mi.
5. **t bin (Mutation)**: Thay i ngu nhin mt s gen  duy tr tnh a dng.
6. **Lp li**: Quay li bc 2 cho n khi t iu kin dng (s th h hoc gi tr fitness mong mun).



### ng dng
- Ti u ha hm s.
- Bi ton lp lch.
- Thit k k thut, machine learning, v.v.

## Cc v d minh ha
Chng ta s thc hin hai v d:

1. Tm gi tr ti a ca hm s $f(x) = -x^2 + 10x + 50$ trong khong $x \in [0, 10]$.
2. Tm gi tr ti thiu ca hm s $g(x, y) = x^2 + y^2$ trong khong $x, y \in [-5, 5]$.

Ngoi ra, chng ta s trc quan ha s thay i ca gi tr fitness qua cc th h.


## V d 1: Ti u ha hm mt bin

### Bi ton  
Tm $x$ sao cho  
$$
f(x) = -(x^2) + 10x + 50
$$  
t gi tr ln nht trong khong  
$$
x \in [0, 10].
$$

# M gi
```plaintext
To qun th mi rng
Lp cho n khi qun th mi  N c th:
    Chn cha m 1 bng Tournament Selection:
        - Chn ngu nhin tournament_size c th trong qun th
        - Chn c th c fitness cao nht lm cha m 1
    Chn cha m 2 bng Tournament Selection tng t
    
    Thc hin lai ghp (crossover) gia cha m 1 v cha m 2 vi xc sut crossover_rate:
        Nu crossover thnh cng:
            To con mi c gi tr trung bnh ca cha m
        Ngc li:
            Con mi l cha m 1
            
    Thc hin t bin cho con mi vi xc sut mutation_rate:
        Nu t bin:
            Gn con mi gi tr ngu nhin trong [min_val, max_val]
            
    Thm con mi vo qun th mi
    
Cp nht qun th hin ti = qun th mi
In ra c th tt nht v gi tr fitness tng ng ca th h


````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def fitness_function(x):
    return -(x**2) + 10*x + 50

# Khi to qun th
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# La chn (Tournament Selection)
def select_parents(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghp
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# t bin
def mutate(individual, mutation_rate=0.1, min_val=0, max_val=10):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thut ton di truyn
def genetic_algorithm_example1(pop_size=50, generations=100, min_val=0, max_val=10):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function(x) for x in population])]
        print(f'Th h {generation + 1}: x = {best_individual:.4f}, f(x) = {best_fitness:.4f}')
    
    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr f(x) tt nht')
    plt.title('S thay i ca f(x) qua cc th h')
    plt.show()
    
    return best_individual, best_fitness

# Chy thut ton
best_x, best_f = genetic_algorithm_example1()
print(f'\nKt qu cui cng: x = {best_x:.4f}, f(x) = {best_f:.4f}')
````

## V d 2: Ti u ha hm hai bin

### Bi ton  
Tm $(x, y)$ sao cho

### M gi 

```plaintext
To qun th mi rng
Lp cho n khi qun th mi  N c th:
    Chn cha m 1 bng Tournament Selection trn qun th hin ti
    Chn cha m 2 bng Tournament Selection trn qun th hin ti
    Thc hin lai ghp (crossover) gia cha m 1 v cha m 2 vi xc sut crossover_rate:
        Nu crossover thnh cng:
            To con mi bng trung bnh gi tr (x, y) ca cha m
        Ngc li:
            Con mi l cha m 1
    Thc hin t bin cho con mi vi xc sut mutation_rate tng bin:
        Nu t bin:
            Gn gi tr ngu nhin trong [min_val, max_val] cho bin tng ng
    Thm con mi vo qun th mi

Cp nht qun th hin ti = qun th mi
In kt qu tt nht th h

````python
# Hm mc tiu
def fitness_function_2d(x, y):
    return -(x**2 + y**2)  # S dng du m v GA tm max

# Khi to qun th
def initialize_population_2d(pop_size, min_val, max_val):
    return [(random.uniform(min_val, max_val), random.uniform(min_val, max_val)) for _ in range(pop_size)]

# La chn
def select_parents_2d(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghp
def crossover_2d(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return ((parent1[0] + parent2[0]) / 2, (parent1[1] + parent2[1]) / 2)
    return parent1

# t bin
def mutate_2d(individual, mutation_rate=0.1, min_val=-5, max_val=5):
    x, y = individual
    if random.random() < mutation_rate:
        x = random.uniform(min_val, max_val)
    if random.random() < mutation_rate:
        y = random.uniform(min_val, max_val)
    return (x, y)

# Thut ton di truyn
def genetic_algorithm_example2(pop_size=50, generations=100, min_val=-5, max_val=5):
    population = initialize_population_2d(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function_2d(x, y) for x, y in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(-best_fitness)  # Chuyn v gi tr thc
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents_2d(population, fitness_values)
            parent2 = select_parents_2d(population, fitness_values)
            child = crossover_2d(parent1, parent2)
            child = mutate_2d(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
        print(f'Th h {generation + 1}: (x, y) = ({best_individual[0]:.4f}, {best_individual[1]:.4f}), g(x, y) = {-best_fitness:.4f}')
    
    best_fitness = max([fitness_function_2d(x, y) for x, y in population])
    best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr g(x, y) tt nht')
    plt.title('S thay i ca g(x, y) qua cc th h')
    plt.show()
    
    return best_individual, -best_fitness

# Chy thut ton
best_xy, best_g = genetic_algorithm_example2()
print(f'\nKt qu cui cng: (x, y) = ({best_xy[0]:.4f}, {best_xy[1]:.4f}), g(x, y) = {best_g:.4f}')
````

# Bi tp 1
# Thay i hm mc tiu**: Th p dng thut ton cho hm $h(x) = sin(x) + cos(x)$ trong khong $x \in [0, 2\pi]$. So snh kt qu vi nghim chnh xc.

## Ti u hm $h(x) = \sin(x) + \cos(x)$ bng Thut Ton Di Truyn (Genetic Algorithm)

## Mc tiu
- Ti u hm:
  $$
  h(x) = \sin(x) + \cos(x)
  $$
  trong min:
  $$
  x \in [0, 2\pi]
  $$
- Tm gi tr **x** sao cho **h(x)** t gi tr ln nht.

---

## Phn tch nghim chnh xc

Ta c:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##  tng thut ton di truyn (GA)

1. **M ha**: Biu din gi tr \( x \) bng chui nh phn di 16 bit.  
2. **Gii m**: Chuyn t chui nh phn sang gi tr thc trong khong \([0, 2\pi]\).  
3. **Fitness**: S dng chnh gi tr \( h(x) \) lm  thch nghi.  
4. **Chn lc**: Roulette Wheel Selection.  
5. **Lai ghp**: Lai 1 im vi xc sut cao.  
6. **t bin**: Lt bit vi xc sut thp.

---


````python
# code vo  y nhe
# Gii Bi tp 1 bng Thut ton Di truyn (Genetic Algorithm)
# -----------------------------------------------------------
# Mc tiu: Ti a ha h(x) = sin(x) + cos(x) trn [0, 2]
# - M ha nh phn 16bit
# - Chn lc Roulette Wheel
# - Lai ghp 1 im
# - t bin lt bit
# -----------------------------------------------------------
import math
import random
import matplotlib.pyplot as plt

# ----------------------- Tham s GA ------------------------
BITS = 16               # s bit m ha
POP_SIZE = 50           # kch thc qun th
GENERATIONS = 100       # s th h
CROSS_RATE = 0.9        # xc sut lai
MUT_RATE = 0.005        # xc sut t bin mi bit
LOW, HIGH = 0.0, 2 * math.pi  # min tm kim

# ------------- Hm mc tiu & tr gip --------------------
def h(x: float) -> float:
    """Hm mc tiu cn ti a ha."""
    return math.sin(x) + math.cos(x)

def encode(x: float) -> int:
    """Chuyn gi tr thc x -> gene nh phn (kiu int)."""
    return int(round((x - LOW) / (HIGH - LOW) * (2**BITS - 1)))

def decode(gene: int) -> float:
    """Chuyn gene nh phn -> gi tr thc x."""
    return LOW + gene / (2**BITS - 1) * (HIGH - LOW)

def random_gene() -> int:
    """Sinh gene ngu nhin."""
    return random.randint(0, 2**BITS - 1)

def gene_fitness(gene: int) -> float:
    """ thch nghi = gi tr hm h(x)."""
    return h(decode(gene))

# -------------------- Cc ton t GA ----------------------
def roulette_select(pop):
    """Chn lc Roulette Wheel  t l thun fitness ( dch >0)."""
    fits = [gene_fitness(g) for g in pop]
    min_fit = min(fits)
    shifted = [f - min_fit + 1e-9 for f in fits]  # m bo >0
    total = sum(shifted)
    r = random.uniform(0, total)
    acc = 0.0
    for g, w in zip(pop, shifted):
        acc += w
        if acc >= r:
            return g
    return pop[-1]

def crossover(g1: int, g2: int):
    """Lai ghp 1 im, tr v 2 con."""
    if random.random() < CROSS_RATE:
        point = random.randint(1, BITS - 1)
        mask = (1 << point) - 1
        child1 = (g1 & ~mask) | (g2 & mask)
        child2 = (g2 & ~mask) | (g1 & mask)
        return child1, child2
    return g1, g2

def mutate(g: int) -> int:
    """t bin: lt ngu nhin cc bit vi xc sut MUT_RATE."""
    for bit in range(BITS):
        if random.random() < MUT_RATE:
            g ^= (1 << bit)
    return g

# ----------------------- Chy GA ---------------------------
random.seed(42)  # c nh seed  ti lp kt qu
population = [random_gene() for _ in range(POP_SIZE)]
best_history = []  # lu gi tr tt nht qua cc th h

for gen in range(1, GENERATIONS + 1):
    best_gene = max(population, key=gene_fitness)   # c th tt nht
    best_x = decode(best_gene)
    best_h = h(best_x)
    
    # In log theo nh dng yu cu
    print(f"Th h {gen}: x = {best_x: .4f}, h(x) = {best_h: .4f}")
    
    best_history.append(best_h)
    
    # To qun th mi
    new_pop = []
    while len(new_pop) < POP_SIZE:
        p1, p2 = roulette_select(population), roulette_select(population)
        c1, c2 = crossover(p1, p2)
        new_pop.extend([mutate(c1), mutate(c2)])
    population = new_pop[:POP_SIZE]

# ----------------- So snh vi nghim chnh xc ------------
exact_x = math.pi / 4
exact_h = math.sqrt(2)
print("\n--- So snh vi nghim chnh xc ---")
print(f"GA:     x* = {best_x:.6f} rad ({math.degrees(best_x):.4f}), h(x*) = {best_h:.6f}")
print(f"Exact:  x* = {exact_x:.6f} rad ({math.degrees(exact_x):.4f}), h(x*) = {exact_h:.6f}")
print(f"x = {abs(best_x - exact_x):.6e} rad, h = {abs(best_h - exact_h):.6e}")

# -------------------- V biu  hi t --------------------
plt.figure(figsize=(8, 4))
plt.plot(range(1, GENERATIONS + 1), best_history, linewidth=2)
plt.title("Biu  hi t GA  h(x) tt nht theo th h")
plt.xlabel("Th h")
plt.ylabel("Gi tr h(x) tt nht")
plt.grid(True)
plt.tight_layout()
plt.show()
````

# Bi tp 2: Ti u hm $h(x) = \sin(x) + \cos(x)$ bng Thut Ton Di Truyn (Genetic Algorithm)

---

## Mc tiu

Ti u hm:
$$
h(x) = \sin(x) + \cos(x)
$$
trong khong:
$$
x \in [0, 2\pi]
$$

> Tm gi tr $x$ sao cho $h(x)$ t **gi tr ln nht**.

---

## Phn tch nghim chnh xc

Ta c:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##   tng thut ton di truyn (GA)

| Thnh phn        | Chi tit                                                                 |
|------------------|-------------------------------------------------------------------------|
| **M ha**        | Dng chui nh phn 16 bit  biu din gi tr $x \in [0, 2\pi]$         |
| **Gii m**       | nh x nh phn sang gi tr thc: $x = 0 + \dfrac{\text{decimal}}{2^{16}-1} \cdot 2\pi$ |
| **Fitness**       | Chnh l $h(x) = \sin(x) + \cos(x)$                                     |
| **Chn lc**      | Roulette Wheel Selection                                                |
| **Lai ghp**      | 1 im ct vi xc sut 0.8                                              |
| **t bin**      | Xc sut 5% (0.05)                                                       |

---

````python
# code vo y nhe
# Bi tp2  Ti u h(x)=sin(x)+cos(x) bng GA
# ------------------------------------------------
# Tham s theo : 16bit, roulette, crossover 1im(p=0.8), mutation 5%
# ------------------------------------------------
import math, random, matplotlib.pyplot as plt

# ---------------- Tham s GA --------------------
BITS       = 16          # s bit m ha
POP_SIZE   = 50          # kch thc qun th
GENERATIONS= 100         # s th h
CROSS_RATE = 0.8         # xc sut lai
MUT_RATE   = 0.05        # xc sut t bin mi bit
LOW, HIGH  = 0.0, 2*math.pi

# --------------- Hm mc tiu -------------------
def h(x: float) -> float:
    return math.sin(x) + math.cos(x)

# -------------- M ha / gii m ----------------
def encode(x: float) -> int:
    return int(round((x - LOW) / (HIGH - LOW) * (2**BITS - 1)))

def decode(gene: int) -> float:
    return LOW + gene / (2**BITS - 1) * (HIGH - LOW)

# --------------  thch nghi -------------------
def fitness(gene: int) -> float:
    return h(decode(gene))

# -------------- Chn lc roulette ---------------
def roulette_select(pop):
    fits = [fitness(g) for g in pop]
    min_fit = min(fits)
    shifted = [f - min_fit + 1e-9 for f in fits]  # dch v >0
    s = sum(shifted)
    r = random.uniform(0, s)
    acc = 0.0
    for g, w in zip(pop, shifted):
        acc += w
        if acc >= r:
            return g
    return pop[-1]

# ---------------- Lai  1 im ------------------
def crossover(g1: int, g2: int):
    if random.random() < CROSS_RATE:
        point = random.randint(1, BITS - 1)
        mask = (1 << point) - 1
        return (g1 & ~mask) | (g2 & mask), (g2 & ~mask) | (g1 & mask)
    return g1, g2

# ----------------- t bin ---------------------
def mutate(g: int) -> int:
    for b in range(BITS):
        if random.random() < MUT_RATE:
            g ^= (1 << b)   # lt bit b
    return g

# --------------- Chy GA ------------------------
random.seed(2025)
pop = [random.randint(0, 2**BITS - 1) for _ in range(POP_SIZE)]
best_history = []

for gen in range(1, GENERATIONS + 1):
    best_gene = max(pop, key=fitness)
    best_x = decode(best_gene)
    best_h = h(best_x)
    print(f"Th h {gen:3d}: x = {best_x: .4f}, h(x) = {best_h: .4f}")
    best_history.append(best_h)
    
    # sinh th h mi
    new_pop = []
    while len(new_pop) < POP_SIZE:
        p1, p2 = roulette_select(pop), roulette_select(pop)
        c1, c2 = crossover(p1, p2)
        new_pop.extend([mutate(c1), mutate(c2)])
    pop = new_pop[:POP_SIZE]

# -------- So snh vi nghim chnh xc ---------
exact_x, exact_h = math.pi/4, math.sqrt(2)
print("\n==> Nghim GA:   x* = %.6f rad (%.4f), h(x*) = %.6f" % (best_x, math.degrees(best_x), best_h))
print("    Nghim chun: x* = %.6f rad (%.4f), h(x*) = %.6f" % (exact_x, math.degrees(exact_x), exact_h))
print("    x = %.3e rad, h = %.3e" % (abs(best_x-exact_x), abs(best_h-exact_h)))

# --------------- Biu  hi t -----------------
plt.figure(figsize=(8,4))
plt.plot(range(1, GENERATIONS+1), best_history, lw=2)
plt.title("Biu  hi t GA  h(x) tt nht qua cc th h")
plt.xlabel("Th h")
plt.ylabel("h(x) tt nht")
plt.grid(True)
plt.tight_layout()
plt.show()
````

# Thay i tham s thut ton di truyn v phn tch tc  hi t

**Mc tiu:** Th thay i cc tham s chnh ca thut ton di truyn gm:

- Kch thc qun th: `pop_size`
- S th h: `generations`
- T l t bin: `mutation_rate`
- T l lai ghp: `crossover_rate`

v phn tch nh hng ca tng tham s n tc  hi t ca thut ton.

---

## Cc tham s v vai tr

| Tham s         |  ngha                                  | nh hng chnh                                                                                         |
|-----------------|-----------------------------------------|--------------------------------------------------------------------------------------------------------|
| `pop_size`      | S c th trong qun th                 | Qun th ln gip khm ph khng gian tm kim tt hn nhng tn nhiu thi gian tnh ton. Qun th nh hi t nhanh nhng d b kt cc tr cc b. |
| `generations`   | S vng lp ca thut ton               | S th h cng nhiu th thut ton c nhiu c hi ci thin nghim, tuy nhin thi gian chy cng tng. |
| `mutation_rate` | Xc sut t bin mi bit                | t bin cao gip tng a dng gene, trnh b kt trong cc tr cc b, nhng nu qu cao lm qu trnh tm kim nhiu lon. |
| `crossover_rate`| Xc sut lai ghp gia hai c th       | Lai ghp gip phi hp gen tt, tng kh nng tm nghim tt. Nu qu thp, qun th km a dng; qu cao c th ph hy cu trc tt ca c th. |

---

## Phn tch nh hng

- **`pop_size`**: Kch thc qun th ln gip tm kim ton cc tt hn nhng chm hn, kch thc nh lm thut ton hi t nhanh nhng d mc kt.
- **`generations`**: S th h ln tng kh nng ci tin kt qu nhng lm tng thi gian tnh ton.
- **`mutation_rate`**: T l t bin va phi gip duy tr a dng di truyn, trnh kt cc b; qu cao hoc qu thp u lm gim hiu qu.
- **`crossover_rate`**: T l lai ghp hp l gip khai thc tt thng tin di truyn, tng hiu qu hi t.

---




````python
# code vo y 
"""
GA Parameter Study  h(x) = sin(x) + cos(x)
- M ha 16-bit, roulette, crossover 1 im, mutation lt bit
- Th nhiu gi tr pop_size, generations, mutation_rate, crossover_rate
- nh gi: th h u t ngng (2 - 1e-4) & h(x) cui
- Hin th bng kt qu (pandas) + biu  Matplotlib
"""
import math, random, pandas as pd, matplotlib.pyplot as plt
from typing import List, Dict

# ---------- Cu hnh & hng s ----------
BITS       = 16
LOW, HIGH  = 0.0, 2*math.pi
DEFAULTS   = dict(pop_size=50, generations=100,
                  crossover_rate=0.8, mutation_rate=0.05)
THRESHOLD  = math.sqrt(2) - 1e-4  # ~1.414114

# ---------- Hm mc tiu & GA ----------
def h(x: float) -> float:
    return math.sin(x) + math.cos(x)

def encode(x: float) -> int:
    return int(round((x-LOW)/(HIGH-LOW)*(2**BITS-1)))

def decode(g: int) -> float:
    return LOW + g/(2**BITS-1)*(HIGH-LOW)

def fitness(g: int) -> float:
    return h(decode(g))

def roulette(pop):
    fits = [fitness(g) for g in pop]
    m = min(fits)
    shifted = [f-m+1e-9 for f in fits]
    total = sum(shifted)
    r = random.uniform(0, total)
    acc = 0
    for g,w in zip(pop, shifted):
        acc += w
        if acc >= r:
            return g
    return pop[-1]

def crossover(g1, g2, rate):
    if random.random() < rate:
        p = random.randint(1, BITS-1)
        mask = (1<<p)-1
        return (g1&~mask)|(g2&mask), (g2&~mask)|(g1&mask)
    return g1, g2

def mutate(g, rate):
    for b in range(BITS):
        if random.random() < rate:
            g ^= (1<<b)
    return g

def run_ga(pop_size, generations, crossover_rate, mutation_rate, seed=0)->List[float]:
    random.seed(seed)
    pop = [random.randint(0, 2**BITS-1) for _ in range(pop_size)]
    best_hist=[]
    for _ in range(generations):
        best_gene = max(pop, key=fitness)
        best_hist.append(fitness(best_gene))
        new=[]
        while len(new) < pop_size:
            p1,p2 = roulette(pop), roulette(pop)
            c1,c2 = crossover(p1,p2,crossover_rate)
            new.extend([mutate(c1,mutation_rate), mutate(c2,mutation_rate)])
        pop = new[:pop_size]
    return best_hist

def first_hit(hist, thr=THRESHOLD):
    for i,v in enumerate(hist,1):
        if v >= thr:
            return i
    return None

# ---------- B cu hnh th nghim ----------
tests=[]
for ps in [20,50,100]:
    cfg=DEFAULTS.copy(); cfg['pop_size']=ps
    tests.append(("pop_size",ps,cfg))
for g in [30,60,100]:
    cfg=DEFAULTS.copy(); cfg['generations']=g
    tests.append(("generations",g,cfg))
for mr in [0.01,0.05,0.1]:
    cfg=DEFAULTS.copy(); cfg['mutation_rate']=mr
    tests.append(("mutation_rate",mr,cfg))
for cr in [0.6,0.8,0.95]:
    cfg=DEFAULTS.copy(); cfg['crossover_rate']=cr
    tests.append(("crossover_rate",cr,cfg))

# ---------- Chy & thu kt qu ----------
records=[]
histories: Dict[str,Dict[float,List[float]]] = {}
for param,val,cfg in tests:
    hist = run_ga(**cfg, seed=2025)
    histories.setdefault(param,{})[val]=hist
    records.append(dict(Parameter=param, Value=val,
                        Gen_hit=first_hit(hist),
                        Best_final=hist[-1]))

df = pd.DataFrame(records)
print("\nBNG KT QU:")
print(df.to_string(index=False))

# ---------- Hm v ----------
def plot_group(name, dic):
    plt.figure(figsize=(8,4))
    for val,hist in dic.items():
        plt.plot(range(1,len(hist)+1), hist, label=str(val))
    plt.title(f"nh hng {name}")
    plt.xlabel("Th h"); plt.ylabel("h(x) tt nht")
    plt.grid(True); plt.legend(); plt.tight_layout(); plt.show()

for group in histories:
    plot_group(group, histories[group])
````

   - Gi tr nghim tt nht t c.

- Phng php **Roulette Wheel** thng gip tng tc  hi t nh u tin c th tt.
- Tuy nhin c th lm gim a dng gen nu c th u t chim u th qu ln, d ri vo cc tr cc b.
- Phng php **Tournament** kim sot tt s a dng nhng c th chm hn trong mt s trng hp.
- Phng php **Random** n gin nhng t hiu qu trong ti u nng cao.

````python
# code vo y nh
"""
BI TP4  So snh ba phng php chn lc GA
==============================================

 Ti u h(x)=sin(x)+cos(x) trn [0,2]  
 Ba phng php chn lc: Random, Roulette, Tournament  
 Tham s c nh: 16bit, pop=50, gen=100, crossover=0.8, mutation=0.05  
 Chy 20 seed  tnh trung bnh:  
   Gen u tin t ngng (21e4)  
   Gi tr h(x) cui  
 V biu  fitness trung bnh theo th h
"""
import math, random, statistics, pandas as pd, matplotlib.pyplot as plt
from typing import List, Dict, Callable

# ------------------ Cu hnh chung ------------------
BITS = 16
LOW, HIGH = 0.0, 2*math.pi
POP_SIZE = 50
GENERATIONS = 100
CROSS_RATE = 0.8
MUT_RATE = 0.05
THRESHOLD = math.sqrt(2) - 1e-4  # ngng hi t
RUNS = 20                         # s ln chy mi phng php

# ------------------ Hm GA c bn -------------------
def h(x): return math.sin(x) + math.cos(x)
def encode(x): return int(round((x-LOW)/(HIGH-LOW)*(2**BITS-1)))
def decode(g): return LOW + g/(2**BITS-1)*(HIGH-LOW)
def fitness(g): return h(decode(g))

# --- Chn lc ---
def sel_random(pop): return random.choice(pop)

def sel_roulette(pop):
    fits = [fitness(g) for g in pop]
    m = min(fits)
    shifted = [f - m + 1e-9 for f in fits]
    total = sum(shifted)
    pick = random.uniform(0,total)
    acc=0
    for g,w in zip(pop,shifted):
        acc += w
        if acc >= pick:
            return g
    return pop[-1]

def sel_tournament(pop, k=3):
    return max(random.sample(pop,k), key=fitness)

# --- Ton t di truyn ---
def crossover(g1,g2):
    if random.random()<CROSS_RATE:
        p=random.randint(1,BITS-1)
        mask=(1<<p)-1
        return (g1&~mask)|(g2&mask), (g2&~mask)|(g1&mask)
    return g1,g2
def mutate(g):
    for b in range(BITS):
        if random.random()<MUT_RATE:
            g ^= (1<<b)
    return g

# --- Hm chy GA mt ln ---
def run_ga(selector:Callable[[List[int]],int], seed:int)->List[float]:
    random.seed(seed)
    pop=[random.randint(0,2**BITS-1) for _ in range(POP_SIZE)]
    best_hist=[]
    for _ in range(GENERATIONS):
        best_hist.append(fitness(max(pop,key=fitness)))
        new=[]
        while len(new)<POP_SIZE:
            p1,p2=selector(pop),selector(pop)
            c1,c2=crossover(p1,p2)
            new.extend([mutate(c1), mutate(c2)])
        pop=new[:POP_SIZE]
    return best_hist

def first_hit(hist, thr=THRESHOLD):
    for i,v in enumerate(hist,1):
        if v>=thr: return i
    return None

methods = {
    "Random":    sel_random,
    "Roulette":  sel_roulette,
    "Tournament":lambda pop: sel_tournament(pop,3)
}

records=[]
avg_histories={}  # mean fitness per gen
for name,selector in methods.items():
    all_runs=[]
    gen_hits=[]
    best_finals=[]
    for seed in range(RUNS):
        hist=run_ga(selector, seed)
        all_runs.append(hist)
        gen_hits.append(first_hit(hist) or GENERATIONS)
        best_finals.append(hist[-1])
    # trung bnh theo th h
    avg_hist=[statistics.mean(col) for col in zip(*all_runs)]
    avg_histories[name]=avg_hist
    records.append({
        "Method":name,
        "AvgGenHit": statistics.mean(gen_hits),
        "StdGenHit": statistics.stdev(gen_hits),
        "AvgFinalh(x)": statistics.mean(best_finals)
    })

# ---------- Bng kt qu ----------
df=pd.DataFrame(records)
print("\nKT QU TRUNG BNH SAU", RUNS, "LN CHY:")
print(df.to_string(index=False))

# ---------- Biu  hi t ----------
plt.figure(figsize=(8,4))
for name,hist in avg_histories.items():
    plt.plot(range(1,GENERATIONS+1), hist, label=name)
plt.title("Fitness trung bnh theo th h (20 ln chy)")
plt.xlabel("Th h")
plt.ylabel("h(x) tt nht (trung bnh)")
plt.grid(True); plt.legend(); plt.tight_layout(); plt.show()
````

# Bi tp 5
# Trc quan ha qun th  V biu  phn tn (scatter plot) th hin s phn b ca qun th qua cc th h

---

##  Mc tiu

- Trc quan ha s phn b ca qun th c th trong qu trnh tin ha ca thut ton di truyn.
- Dng biu  phn tn (scatter plot)  th hin v tr cc c th trn khng gian bin s qua tng th h.
- Gip quan st r tin trnh hi t, a dng di truyn, v s thay i qun th theo thi gian.

---

##  tng thc hin

- Vi bi ton ti u hm $k(x, y, z) = x^2 + y^2 + z^2$, khng gian tm kim l 3 chiu.
- Mi c th l mt im trong khng gian $(x, y, z)$.
-  mi th h, ly v tr gii m ca ton b c th, v scatter plot.
- Nu khng th trc quan 3D trc tip, c th v scatter 2D tng cp bin: $(x, y)$, $(y, z)$, $(x, z)$ qua cc th h.

---

````python
#code vo y
import math, random, matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ------------------- GA parameters -------------------
POP_SIZE = 80
GENERATIONS = 100
CROSS_RATE = 0.8
MUT_RATE = 0.1
LOW, HIGH = -5.0, 5.0  # search space for each dimension (x,y,z)

# ------------------- Objective function --------------
def k(x, y, z):
    """k(x,y,z) = x^2 + y^2 + z^2 (minimize -> optimum at (0,0,0))."""
    return x**2 + y**2 + z**2

# ---------------- Individual helpers -----------------
def random_individual():
    return [random.uniform(LOW, HIGH) for _ in range(3)]

def fitness(ind):
    # negate because we want to maximise in GA framework
    return -k(*ind)

def mutate(ind):
    for i in range(3):
        if random.random() < MUT_RATE:
            ind[i] += random.gauss(0, 0.5)
            ind[i] = max(LOW, min(HIGH, ind[i]))
    return ind

def crossover(p1, p2):
    if random.random() < CROSS_RATE:
        alpha = random.random()
        return [
            alpha * p1[i] + (1 - alpha) * p2[i] for i in range(3)
        ], [
            (1 - alpha) * p1[i] + alpha * p2[i] for i in range(3)
        ]
    return p1, p2

def tournament_select(pop, k=3):
    return max(random.sample(pop, k), key=fitness)

# ------------------- GA run with logging -------------
random.seed(1)
population = [random_individual() for _ in range(POP_SIZE)]
pop_history = [population.copy()]

for gen in range(1, GENERATIONS + 1):
    new_population = []
    while len(new_population) < POP_SIZE:
        p1 = tournament_select(population)
        p2 = tournament_select(population)
        c1, c2 = crossover(p1, p2)
        new_population.extend([mutate(c1), mutate(c2)])
    population = new_population[:POP_SIZE]
    pop_history.append(population.copy())

# --------------- Visualization (scatter) -------------
def plot_generation(pop, gen_idx):
    xs = [p[0] for p in pop]
    ys = [p[1] for p in pop]
    zs = [p[2] for p in pop]
    
    fig = plt.figure(figsize=(6,5))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(xs, ys, zs)
    ax.set_title(f"Phn b qun th  Th h {gen_idx}")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")
    plt.tight_layout()
    plt.show()

# Plot for th h 0, 10, 50, 100
for idx in [0, 10, 50, 100]:
    plot_generation(pop_history[idx], idx)
````
